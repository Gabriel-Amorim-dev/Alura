`````python
print('Informa√ß√µes Dispon√≠veis:')
print(df.columns.tolist())
info = input('Qual informa√ß√£o voc√™ quer saber? ')

def operations(coluna):
  print('\nEscolha a opera√ß√£o que deseja realizar:')
  print('1 - Ver primeiras linhas\n')
  print('2 - Ver √∫ltimas linhas\n')
  print('3 - Ver tipo de dado e nulos\n')
  print('4 - Ver estat√≠sticas\n')
  print('5 - Ver quantos valores nulos\n')
  print('6 - Ver quantas duplicadas\n')
  print('7 - Ver quantos valores\n')
  print('\n')


  func = int(input('Digite o n√∫mero da opera√ß√£o: '))

  match func:
    case 1:
      print(df[coluna].head())
    case 2:
      print(df[coluna].tail())
    case 3:
      print(f"Tipo de dado: {df[coluna].dtype}")
      print(f"Quantidade de nulos: {df[coluna].isnull().sum()}")
    case 4:
      print(df[coluna].describe())
    case 5:
      print(f"Valores nulos: {df[coluna].isnull().sum()}")
    case 6:
      print(f"Duplicadas: {df[coluna].duplicated().sum()}")
    case 7:
      print(f'Contagem de Valores: {df[coluna].value_counts()}')
    case _:
      print("Op√ß√£o inv√°lida.")

operations(info)
`````

`````python
import pandas as pd
import numpy as np

max_limit = 20

def limite(prompt, limit):
    while True:
      try:
          value = int(input(prompt))
          if 1 <= value <= limit:
              return value
          else:
              print(f"Digite um valor de 1 at√© {limit}.")
      except ValueError:
          print("Entrada inv√°lida. Por favor, digite um n√∫mero inteiro.")

num_linhas = limite('Quantas linhas ter√° seu DataFrame (M√°x 20):', max_limit)
num_colunas = limite('Quantas colunas ter√° seu DataFrame (M√°x 20):', max_limit)

colunas = []
for i in range (num_colunas):
          nome = input(f'Digite o nome da coluna {i+1}: ')
          colunas.append(nome)

dados = []
for i in range (num_linhas):
      linha = []
      print(f"\n--- Preenchendo linha {i+1} ---")
      for col in colunas:
          valor = input(f'Insira o valor aqui para a coluna "{col}" (Ou digite NaN para deixar v√°zio): ')
          if valor.lower() == 'nan':
              linha.append(np.nan)
          else:
              linha.append(valor)

      dados.append(linha)

DFN = pd.DataFrame(dados, columns=colunas)

print("\nDataFrame criado:")
print(DFN)
`````
