`````python
print('Informações Disponíveis:')
print(df.columns.tolist())
info = input('Qual informação você quer saber? ')

def operations(coluna):
  print('\nEscolha a operação que deseja realizar:')
  print('1 - Ver primeiras linhas')
  print('2 - Ver últimas linhas')
  print('3 - Ver tipo de dado e nulos')
  print('4 - Ver estatísticas')
  print('5 - Ver quantos valores nulos')
  print('6 - Ver quantas duplicadas')
  print('7 - Ver quantos valores')


  func = int(input('Digite o número da operação: '))

  match func:
    case 1:
      print(df[coluna].head())
    case 2:
      print(df[coluna].tail())
    case 3:
      print(f"Tipo de dado: {df[coluna].dtype}")
      print(f"Quantidade de nulos: {df[coluna].isnull().sum()}")
    case 4:
      print(df[coluna].describe())
    case 5:
      print(f"Valores nulos: {df[coluna].isnull().sum()}")
    case 6:
      print(f"Duplicadas: {df[coluna].duplicated().sum()}")
    case 7:
      print(f'Contagem de Valores: {df[coluna].value_counts()}')
    case _:
      print("Opção inválida.")

operations(info)
`````

---

### 🎯 O que cada gráfico faz:

| Opção | Gráfico                        | Descrição                                                                                              |
| ----- | ------------------------------ | ------------------------------------------------------------------------------------------------------ |
| 1     | **Gráfico de Barras**          | Compara a média da variável numérica para cada categoria, mostrando barras ordenadas.                  |
| 2     | **Gráfico de Pizza**           | Mostra a proporção das 10 maiores categorias, útil para poucas categorias.                             |
| 3     | **Gráfico de Caixa (Boxplot)** | Exibe a distribuição, quartis e possíveis outliers da variável numérica para cada categoria.           |
| 4     | **Gráfico de Linhas**          | Mostra a tendência da média da variável numérica ao longo das categorias (útil para séries temporais). |
| 5     | **Scatterplot**                | Mostra a dispersão dos dados individuais da variável numérica em relação às categorias.                |

---

### 🖥️ Como o código gera os gráficos:

* Cada opção chama a função específica do Matplotlib/Seaborn para desenhar o gráfico com as configurações de títulos, rótulos e tamanhos adequados.
* Gráficos são exibidos em janela de visualização ou notebook.
* O código trata entrada inválida de forma simples.

---

### *✅ Vantagens:*

* Permite uma exploração visual rápida e personalizada de dados.
* Facilita a identificação de padrões, tendências e anomalias.
* Pode ser usado em notebooks para análises interativas.
* Com poucos comandos, gera gráficos claros e informativos.

`````python
import matplotlib.pyplot as plt
import seaborn as sns

# Mostra as colunas
print('Informações Disponíveis:')
print(df.columns.tolist())

try:
    info_categoria = input('Qual coluna categórica você quer analisar? ')
    info_valor = input('Qual valor numérico você quer visualizar (ex: Salário (USD)? ')

    # Verifica se as colunas existem
    if info_categoria not in df.columns or info_valor not in df.columns:
        raise ValueError("Coluna não encontrada.")

    # Verifica se a coluna de valor é numérica
    if not pd.api.types.is_numeric_dtype(df[info_valor]):
        raise TypeError("A variável inserida não é numérica!")

    # Agrupamento
    info_calc = df.groupby(info_categoria)[info_valor].mean().sort_values(ascending=False)

except Exception as e:
    print(f'Erro: {e}')
else:
    def graficos():
        print('\nEscolha o gráfico que deseja analisar:')
        print('1 - Gráfico de Barras')
        print('2 - Gráfico de Pizza (só para poucos itens)')
        print('3 - Gráfico de Caixa (Boxplot)')
        print('4 - Gráfico de Linhas')
        print('5 - Scatterplot')

        try:
            func = int(input('Digite o número da operação: '))
        except ValueError:
            print("Entrada inválida. Digite um número.")
            return

        match func:
            case 1:
                plt.figure(figsize=(10, 6))
                sns.barplot(x=info_calc.values, y=info_calc.index)
                plt.xlabel(f'Média de {info_valor}')
                plt.ylabel(info_categoria)
                plt.title(f'Média de {info_valor} por {info_categoria}')
                plt.tight_layout()
                plt.show()

            case 2:
                top10 = info_calc.head(10)
                plt.figure(figsize=(8, 8))
                plt.pie(top10, labels=top10.index, autopct='%1.1f%%')
                plt.title(f'Top 10 {info_categoria} por média de {info_valor}')
                plt.show()

            case 3:
                plt.figure(figsize=(10, 6))
                sns.boxplot(x=info_categoria, y=info_valor, data=df)
                plt.xticks(rotation=45)
                plt.title(f'Distribuição de {info_valor} por {info_categoria}')
                plt.tight_layout()
                plt.show()

            case 4:
                plt.figure(figsize=(12, 6))
                sns.lineplot(x=info_calc.index, y=info_calc.values, marker='o')
                plt.title(f'Tendência da média de {info_valor} por {info_categoria}')
                plt.ylabel(f'{info_valor}')
                plt.xlabel(info_categoria)
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

            case 5:
                plt.figure(figsize=(10, 6))
                sns.scatterplot(data=df, x=info_categoria, y=info_valor)
                plt.title(f'Dispersão de {info_valor} por {info_categoria}')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

            case _:
                print("Opção inválida.")

    graficos()
`````
