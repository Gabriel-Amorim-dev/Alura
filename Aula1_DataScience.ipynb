### üéØ O que cada gr√°fico faz:

| Op√ß√£o | Gr√°fico                        | Descri√ß√£o                                                                                              |
| ----- | ------------------------------ | ------------------------------------------------------------------------------------------------------ |
| 1     | **Gr√°fico de Barras**          | Compara a m√©dia da vari√°vel num√©rica para cada categoria, mostrando barras ordenadas.                  |
| 2     | **Gr√°fico de Pizza**           | Mostra a propor√ß√£o das 10 maiores categorias, √∫til para poucas categorias.                             |
| 3     | **Gr√°fico de Caixa (Boxplot)** | Exibe a distribui√ß√£o, quartis e poss√≠veis outliers da vari√°vel num√©rica para cada categoria.           |
| 4     | **Gr√°fico de Linhas**          | Mostra a tend√™ncia da m√©dia da vari√°vel num√©rica ao longo das categorias (√∫til para s√©ries temporais). |
| 5     | **Scatterplot**                | Mostra a dispers√£o dos dados individuais da vari√°vel num√©rica em rela√ß√£o √†s categorias.                |

---

### üñ•Ô∏è Como o c√≥digo gera os gr√°ficos:

* Cada op√ß√£o chama a fun√ß√£o espec√≠fica do Matplotlib/Seaborn para desenhar o gr√°fico com as configura√ß√µes de t√≠tulos, r√≥tulos e tamanhos adequados.
* Gr√°ficos s√£o exibidos em janela de visualiza√ß√£o ou notebook.

### *‚úÖ Vantagens:*

* Permite uma explora√ß√£o visual r√°pida e personalizada de dados.
* Facilita a identifica√ß√£o de padr√µes, tend√™ncias e anomalias.
* Pode ser usado em notebooks para an√°lises interativas.
* Com poucos comandos, gera gr√°ficos claros e informativos.

`````python
import matplotlib.pyplot as plt
import seaborn as sns

# Mostra as colunas
print('Informa√ß√µes Dispon√≠veis:')
print(df.columns.tolist())

try:
    info_categoria = input('Qual coluna categ√≥rica voc√™ quer analisar? ')
    info_valor = input('Qual valor num√©rico voc√™ quer visualizar (ex: salary_in_usd)? ')

    # Verifica se as colunas existem
    if info_categoria not in df.columns or info_valor not in df.columns:
        raise ValueError("Coluna n√£o encontrada.")

    # Verifica se a coluna de valor √© num√©rica
    if not pd.api.types.is_numeric_dtype(df[info_valor]):
        raise TypeError("A vari√°vel inserida n√£o √© num√©rica!")

    # Agrupamento
    info_calc = df.groupby(info_categoria)[info_valor].mean().sort_values(ascending=False)

except Exception as e:
    print(f'Erro: {e}')
else:
    def graficos():
        print('\nEscolha o gr√°fico que deseja analisar:')
        print('1 - Gr√°fico de Barras')
        print('2 - Gr√°fico de Pizza (s√≥ para poucos itens)')
        print('3 - Gr√°fico de Caixa (Boxplot)')
        print('4 - Gr√°fico de Linhas')
        print('5 - Scatterplot')

        try:
            func = int(input('Digite o n√∫mero da opera√ß√£o: '))
        except ValueError:
            print("Entrada inv√°lida. Digite um n√∫mero.")
            return

        match func:
            case 1:
                plt.figure(figsize=(10, 6))
                sns.barplot(x=info_calc.values, y=info_calc.index)
                plt.xlabel(f'M√©dia de {info_valor}')
                plt.ylabel(info_categoria)
                plt.title(f'M√©dia de {info_valor} por {info_categoria}')
                plt.tight_layout()
                plt.show()

            case 2:
                top10 = info_calc.head(10)
                plt.figure(figsize=(8, 8))
                plt.pie(top10, labels=top10.index, autopct='%1.1f%%')
                plt.title(f'Top 10 {info_categoria} por m√©dia de {info_valor}')
                plt.show()

            case 3:
                plt.figure(figsize=(10, 6))
                sns.boxplot(x=info_categoria, y=info_valor, data=df)
                plt.xticks(rotation=45)
                plt.title(f'Distribui√ß√£o de {info_valor} por {info_categoria}')
                plt.tight_layout()
                plt.show()

            case 4:
                plt.figure(figsize=(12, 6))
                sns.lineplot(x=info_calc.index, y=info_calc.values, marker='o')
                plt.title(f'Tend√™ncia da m√©dia de {info_valor} por {info_categoria}')
                plt.ylabel(f'{info_valor}')
                plt.xlabel(info_categoria)
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

            case 5:
                plt.figure(figsize=(10, 6))
                sns.scatterplot(data=df, x=info_categoria, y=info_valor)
                plt.title(f'Dispers√£o de {info_valor} por {info_categoria}')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

            case _:
                print("Op√ß√£o inv√°lida.")

    graficos()
`````
