| 1     | **Gráfico de Barras**          | Compara a média da variável numérica para cada categoria, mostrando barras ordenadas.                  |
| 2     | **Gráfico de Pizza**           | Mostra a proporção das 10 maiores categorias, útil para poucas categorias.                             |

`````python
import matplotlib.pyplot as plt
import seaborn as sns

# Mostra as colunas
print('Informações Disponíveis:')
print(df.columns.tolist())

try:
    info_categoria = input('Qual coluna categórica você quer analisar? ')
    info_valor = input('Qual valor numérico você quer visualizar (ex: salary_in_usd)? ')

    # Verifica se as colunas existem
    if info_categoria not in df.columns or info_valor not in df.columns:
        raise ValueError("Coluna não encontrada.")

    # Verifica se a coluna de valor é numérica
    if not pd.api.types.is_numeric_dtype(df[info_valor]):
        raise TypeError("A variável inserida não é numérica!")

    # Agrupamento
    info_calc = df.groupby(info_categoria)[info_valor].mean().sort_values(ascending=False)

except Exception as e:
    print(f'Erro: {e}')
else:
    def graficos():
        print('\nEscolha o gráfico que deseja analisar:')
        print('1 - Gráfico de Barras')
        print('2 - Gráfico de Pizza (só para poucos itens)')
        print('3 - Gráfico de Caixa (Boxplot)')
        print('4 - Gráfico de Linhas')
        print('5 - Scatterplot')

        try:
            func = int(input('Digite o número da operação: '))
        except ValueError:
            print("Entrada inválida. Digite um número.")
            return

        match func:
            case 1:
                plt.figure(figsize=(10, 6))
                sns.barplot(x=info_calc.values, y=info_calc.index)
                plt.xlabel(f'Média de {info_valor}')
                plt.ylabel(info_categoria)
                plt.title(f'Média de {info_valor} por {info_categoria}')
                plt.tight_layout()
                plt.show()

            case 2:
                top10 = info_calc.head(10)
                plt.figure(figsize=(8, 8))
                plt.pie(top10, labels=top10.index, autopct='%1.1f%%')
                plt.title(f'Top 10 {info_categoria} por média de {info_valor}')
                plt.show()

            case 3:
                plt.figure(figsize=(10, 6))
                sns.boxplot(x=info_categoria, y=info_valor, data=df)
                plt.xticks(rotation=45)
                plt.title(f'Distribuição de {info_valor} por {info_categoria}')
                plt.tight_layout()
                plt.show()

            case 4:
                plt.figure(figsize=(12, 6))
                sns.lineplot(x=info_calc.index, y=info_calc.values, marker='o')
                plt.title(f'Tendência da média de {info_valor} por {info_categoria}')
                plt.ylabel(f'{info_valor}')
                plt.xlabel(info_categoria)
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

            case 5:
                plt.figure(figsize=(10, 6))
                sns.scatterplot(data=df, x=info_categoria, y=info_valor)
                plt.title(f'Dispersão de {info_valor} por {info_categoria}')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

            case _:
                print("Opção inválida.")

    graficos()
`````

###**Identificando Valores Nulos no DataFrame**
---

Esse comando verifica, para cada célula do DataFrame, se o valor é nulo (ausente). Ele retorna um DataFrame booleano de mesma dimensão, onde cada posição indica se o valor correspondente é nulo (`True`) ou não (`False`).

---

### 🔍 O que o código faz:

```python
df.isnull()
```

* Analisa todas as colunas e linhas do DataFrame.
* Retorna um novo DataFrame com valores booleanos (`True` para valores nulos e `False` para valores não nulos).
* Permite identificar exatamente onde estão os dados faltantes.

---

### ✅ Vantagens:

* Facilita a visualização e localização de valores ausentes nos dados.
* Pode ser usado para análises e operações de limpeza, como remoção ou preenchimento de dados nulos.

###**Criação Interativa de DataFrame Personalizado**

### ✅ Vantagens:

* Permite criar um DataFrame personalizado, ajustado às necessidades do usuário.
* Facilita a inclusão manual de dados sem precisar criar arquivos externos.
* Suporta a criação de valores ausentes para simular dados reais incompletos.
* Útil para testes, experimentos ou para aprender manipulação de dados no Pandas.


`````python
import pandas as pd
import numpy as np

max_limit = 20

def limite(prompt, limit):
    while True:
      try:
          value = int(input(prompt))
          if 1 <= value <= limit:
              return value
          else:
              print(f"Digite um valor de 1 até {limit}.")
      except ValueError:
          print("Entrada inválida. Por favor, digite um número inteiro.")

num_linhas = limite('Quantas linhas terá seu DataFrame (Máx 20):', max_limit)
num_colunas = limite('Quantas colunas terá seu DataFrame (Máx 20):', max_limit)

colunas = []
for i in range (num_colunas):
          nome = input(f'Digite o nome da coluna {i+1}: ')
          colunas.append(nome)

dados = []
for i in range (num_linhas):
      linha = []
      print(f"\n--- Preenchendo linha {i+1} ---")
      for col in colunas:
          valor = input(f'Insira o valor aqui para a coluna "{col}" (Ou digite NaN para deixar vázio): ')
          if valor.lower() == 'nan':
              linha.append(np.nan)
          else:
              linha.append(valor)

      dados.append(linha)

DFN = pd.DataFrame(dados, columns=colunas)

print("\nDataFrame criado:")
print(DFN)
`````
